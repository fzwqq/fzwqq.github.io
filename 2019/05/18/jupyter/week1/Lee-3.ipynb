```python
train_x = []
train_y = []

for i in range(15):
    x = train_data.iloc[:,i:i+9]
    x.columns=np.array(range(9)) # if we not set the cloumns name, it will have diff cloumns name in every iteration
    y = train_data.iloc[:,i+9]
    y.cloumns = np.array(range(1)) 
    train_x.append(x)
    train_y.append(y)

#   train_y.append(y)
```

```python
train_x = pd.concat(train_x)
train_y = pd.concat(train_y)
# print(train_y)
```

```python
train_x = np.array(train_x,float)
train_y = np.array(train_y,float)
train_x.shape
train_x = np.concatenate((np.ones((train_x.shape[0],1)), train_x), axis = 1)
train_x
```

1. $y = X * w + b$
2. loss = $\frac{1}{2}((y-\hat{y})^2)$
3. $$ \frac{dy}{dw_i} = (y - \hat{y})*x_i $$

```python
w = np.random.random(train_x.shape[1])
# w = np.zeros(len())
lr = 1e-1
epoches = 10000
sum_gra = np.zeros(train_x.shape[1])
for i in range(epoches):
    y_hat = np.dot(train_x,w)
#     loss = 1./2 * ( y_hat - train_y) ** 2
    loss = y_hat - train_y
    gra = 2*np.dot(train_x.transpose(),loss)
#     print(train_x.shape, gra.shape,loss.shape,y_hat.shape,train_y.shape)
    sum_gra += gra**2
    ada = np.sqrt(sum_gra)
    w -= lr * gra / ada
```
